### Lets drop Normalized losses as it comprises 20 percent of non useful data

```python
raw_dataframe.drop('normalized-losses',inplace=True,axis=1)
```

```python
# dropping the "?"s
print(raw_dataframe.shape)
raw_dataframe = raw_dataframe[raw_dataframe['num-of-doors'] != '?']
raw_dataframe = raw_dataframe[raw_dataframe['bore'] != '?']
raw_dataframe = raw_dataframe[raw_dataframe['stroke'] != '?']
raw_dataframe = raw_dataframe[raw_dataframe['horsepower'] != '?']
raw_dataframe = raw_dataframe[raw_dataframe['peak-rpm'] != '?']
raw_dataframe = raw_dataframe[raw_dataframe['price'] != '?']
print(raw_dataframe.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
raw_categorial = raw_dataframe.select_dtypes(include=['object'])
```

```python
le = LabelEncoder()
raw_categorial = raw_categorial.apply(le.fit_transform)
raw_categorial
```

```python
# concat raw_categorial with original df
raw_dataframe = raw_dataframe.drop(raw_categorial.columns, axis=1)
raw_dataframe = pd.concat([raw_dataframe, raw_categorial], axis=1)
raw_dataframe.head()
```

# Train Test Split

```python
x = raw_dataframe.drop(['price'], axis =1)
y = raw_dataframe['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

# Train Model

```python
reg = DecisionTreeRegressor(max_depth=5)
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred_train = reg.predict(X_train)
```

```python
y_pred_test = reg.predict(X_test)
```

### Training Accuracy : 

#### Rsuqred training

```python
reg.score(X_train,y_train)
```

```python
r2_score(y_train,y_pred_train)
```

#### Rsuqred Test

```python
reg.score(X_test,y_test)
```

```python
r2_score(y_test,y_pred_test)
```

#### Residuals

```python
mean_absolute_error(y_test,y_pred_test)
```

```python
mean_squared_error(y_test,y_pred_test)
```

- min_samples_split,

```python
params = {
    'criterion' :["squared_error", "friedman_mse", "absolute_error", "poisson"],
    #'splitter' : [ "best", "random" ],
    'max_depth' : range(3, 10),
    'min_samples_split' : range(5,100,20),
    'min_samples_leaf' : range(5,100,20)   
}
```

# Tunning Individual Params :

```python
# GridSearchCV to find optimal max_depth
from sklearn.model_selection import GridSearchCV


# specify number of folds for k-fold CV
n_folds = 5

# parameters to build the model on
parameters = {'max_depth': range(1, 10)}

# instantiate the model
dtree = DecisionTreeRegressor(random_state =5)

# fit tree on training data
tree = GridSearchCV(dtree, parameters, 
                    cv=n_folds,
                    return_train_score=True)
tree.fit(X_train, y_train)
```

```python
# scores of GridSearch CV
scores = tree.cv_results_
pd.DataFrame(scores)
```

```python
# plotting accuracies with max_depth
plt.figure()
plt.plot(scores["param_max_depth"], 
         scores["mean_train_score"], 
         label="training accuracy")


plt.plot(scores["param_max_depth"], 
         scores["mean_test_score"], 
         label="test accuracy")
plt.xlabel("max_depth")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
train_scores = [ ]
test_scores = [ ]

for i in range(1,10):
    dt = DecisionTreeRegressor(max_depth=i)
    dt.fit(X_train,y_train)
    train_scores.append(dt.score(X_train,y_train))
    test_scores.append(dt.score(X_test,y_test))

print(train_scores)
print()
print(test_scores)
```
